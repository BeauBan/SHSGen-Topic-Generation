Train

```python
def train(filename, num_topics):
    topic_model = BERTopic(nr_topics=num_topics)
    temp_texts = open(filename).read().splitlines()
    texts = []
    for temp_text in temp_texts:
      if temp_text == '': 
        continue
      text = re.sub(r'[^a-zA-Z]', ' ', temp_text)
      words = text.lower().split()  # all lowercase
      eng_stopwords = {}.fromkeys([line.rstrip() for line in stopwords]) # load stop words
      words = [w for w in words if w not in eng_stopwords] # remove stop words
      text = ' '.join(words)
      texts.append(text)

    print('modeling')
    topics, probs = topic_model.fit_transform(texts)
    print(topic_model.get_topic_info())
    model_path = '/content/drive/MyDrive/ColabDrive/Bertopic/result/' + os.path.splitext(os.path.basename(filename))[0] + '_' + 'topic' + str(num_topics) + '.model'
    topic_model.save(model_path)
    return model_path

```

```python
def find_topic_keyterm(keyterm):
  topic_num, similarity = topic_model.find_topics(keyterm, top_n=5)
  representative_docs = topic_model.get_representative_docs(topic_num[0])
  topic_keyterm = topic_model.get_topic(topic_num[0])

  return topic_num[0], representative_docs, topic_keyterm
```
