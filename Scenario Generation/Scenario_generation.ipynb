Prepare prefix, execute once

```python
import gpt_2_simple as gpt2
import os, sys

working_dir =  "/content/drive/MyDrive/ColabDrive/gpt-2-simple-master/"
os.chdir(working_dir)

data_file_path='/content/drive/MyDrive/ColabDrive/gpt-2-simple-master/data/IoTAppSeqWithSymbol.txt'
# prefix="<|startoftext|>Security camera turns on,"

temp_list=[]

with open(data_file_path, 'r') as data_read:
  temp_list = data_read.readlines()

prefix_list=[]
for chain in temp_list:
  if chain.find(',') > 0:  # if there is ',' in the rule chain
    first_rule=chain[:chain.find(',') + 1] # if contains ',', so use the first rule
  else:
    first_rule = chain # if not, return whole rule chain
  # prefix_list.append('<|startoftext|>' + first_rule)
  prefix_list.append(first_rule)

print(prefix_list)
  
```
