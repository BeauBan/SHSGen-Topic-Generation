#Code

Fine tune the model. The parameter for fine tune can be found in [here](https://github.com/minimaxir/gpt-2-simple/blob/master/gpt_2_simple/gpt_2.py)

```python
import gpt_2_simple as gpt2
import os, sys
import requests

working_dir =  "/content/drive/MyDrive/ColabDrive/gpt-2-simple-master/"
os.chdir(working_dir)


# Parameters
# You could specified the parameters. 
# If the line is commented, the parameterts are default.
file_name=os.path.join(working_dir, "data/sceanrio.txt")
model_name="355M"
steps = 100000
# model_dir='models',
# combine=50000,
# batch_size=1,
# learning_rate=0.0001,
# accumulate_gradients=5,
# restore_from='latest',
run_name='fine_tune_355M_scen+smartthings+ifttt_run1'
checkpoint_dir=os.path.join(working_dir,'checkpoint')
# sample_every=100,
sample_length=95
# sample_num=1,
# multi_gpu=False,
save_every=5000
# print_every=1,
# max_checkpoints=1,
# use_memory_saving_gradients=False,
# only_train_transformer_layers=False,
# optimizer='adam',
# overwrite=False):
# See definition at the bottom of the file


sess = gpt2.start_tf_sess()
print("training start")
gpt2.finetune(sess,
              file_name,
              model_name=model_name,
              steps=steps,
              save_every=save_every,
              run_name=run_name,
              checkpoint_dir=checkpoint_dir,
              sample_length=sample_length) 


```
